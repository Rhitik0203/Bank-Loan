```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loan_data = pd.read_csv('financial_loan.csv')
loan_data.head()
```

Basic Data Understanding

```python
loan_data.info()
```

```python
loan_data.describe()
```

Handling Missing Values

```python
loan_data.isnull().sum()
```

```python
loan_data = loan_data.dropna(subset=['emp_title'])
```

Exploratory Data Analysis (EDA)

```python
#Distribution of Key Numerical Variables
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
sns.histplot(loan_data['annual_income'], bins=50, kde=True)
plt.title('Annual Income Distribution')

plt.subplot(2, 2, 2)
sns.histplot(loan_data['loan_amount'], bins=50, kde=True)
plt.title('Loan Amount Distribution')

plt.subplot(2, 2, 3)
sns.histplot(loan_data['int_rate'], bins=30, kde=True)
plt.title('Interest Rate Distribution')

plt.subplot(2, 2, 4)
sns.histplot(loan_data['dti'], bins=30, kde=True)
plt.title('Debt-to-Income Ratio Distribution')

plt.tight_layout()
plt.show()
```

```python
#Categorical Variables Distribution
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='grade', data=loan_data, palette="viridis")
plt.title('Loan Grade Distribution')

plt.subplot(1, 2, 2)
sns.countplot(x='loan_status', data=loan_data, palette="magma")
plt.title('Loan Status Distribution')

plt.tight_layout()
plt.show()
```

Correlation Analysis

```python
# Check for non-numeric columns
print(loan_data.select_dtypes(exclude=['number']).columns)
```

```python
# Select only numeric columns for correlation
numeric_data = loan_data.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_data.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Key Numerical Variables")
plt.show()
```

Additional Analysis

```python
# Analyzing average loan amount and interest rate per loan grade
loan_grade_summary = loan_data.groupby('grade')[['loan_amount', 'int_rate']].mean()
print("\nAverage Loan Amount and Interest Rate by Grade:")
print(loan_grade_summary)
```

```python
# Analyzing loan status distribution across loan purposes
loan_status_purpose = loan_data.groupby(['purpose', 'loan_status']).size().unstack(fill_value=0)
print("\nLoan Status by Purpose:")
print(loan_status_purpose)
```
